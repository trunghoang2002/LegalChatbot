```python
from handle_graph import retrieve
import time

questions = [
    "Điều 5 của Bộ luật Lao động quy định gì?",
    "Người lao động có được nghỉ thai sản bao nhiêu ngày?",
    "Lừa đảo chiếm đoạt tài sản bị phạt bao nhiêu năm tù?",
    "Trộm cáp tài sản trị giá 100 triệu đồng bị phạt bao nhiêu năm tù?",
]
list_state = [
    {"question": questions[i]} for i in range(len(questions))
]

all_time = 0
for state in list_state:
    start_time = time.time()
    state = retrieve(state)
    end_time = time.time()
    all_time += end_time - start_time

print(f"Time average taken: {all_time / len(list_state)} seconds")
```
